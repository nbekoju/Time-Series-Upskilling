### Column Description
Prices (EUR/MWh): Cost of electricity per megawatt-hour, influencing economic factors and consumer demand.

Datetime: Time and date, affecting energy demand patterns based on daily and seasonal variations.

Demand (MW): Total power consumption, directly impacting energy demand.

Temperature: Ambient temperature affecting heating or cooling needs, influencing energy demand.

Feelslike: Perceived temperature, impacting consumer behavior and energy usage patterns.

Dewpoint: Higher dewpoint means the air contains more moisture, while lower dewpoint indicates drier air. Temperature at which air becomes saturated, influencing humidity and potentially affecting cooling systems.

Humidity: Amount of moisture in the air, affecting cooling systems and energy demand.

Precipitation: Rain or snowfall, impacting outdoor activities and potentially affecting energy usage. A high value of precipitation indicates that a large amount of water, in the form of rain, snow, sleet, or other types, has fallen in a specific area over a certain period.

Precipprob: Probability of precipitation, influencing consumer behavior and energy demand planning.

Preciptype: Type of precipitation (rain, snow), impacting energy demand for heating or cooling.

Snow: Snowfall, affecting transportation and potentially influencing energy demand.

Snowdepth: Depth of accumulated snow, influencing outdoor activities and potentially affecting energy demand.

Windgust: Sudden increase in wind speed, impacting energy generation and, indirectly, demand.

Windspeed: Continuous wind speed, affecting energy generation and potentially influencing demand.

Winddirection: Direction from which the wind is blowing, impacting energy generation and demand patterns.

Sealevelpressure: Atmospheric pressure, affecting weather patterns and potentially influencing energy demand.

Cloudcover: Fraction of sky covered by clouds, influencing solar power generation and energy demand.

Visibility: Horizontal distance an observer can see, affecting transportation and potentially influencing energy demand.

Solarradiation: Sun's energy reaching Earth's surface, influencing solar power generation and energy demand.

UVindex: Measure of ultraviolet radiation, potentially influencing outdoor activities and energy demand.

Severerisk: Likelihood of severe weather events, impacting infrastructure and potentially affecting energy demand.

Conditions: Overall weather state, including multiple factors, impacting consumer behavior and energy usage patterns.

```python
for i, feature in enumerate(categorical_features):
  plt.figure(figsize = (10, 8))
  value_counts = train_data[feature].value_counts()

  # create bar plot
  bars = sns.barplot(x = value_counts.index, y = value_counts.values)
  plt.title(f'{feature} - bar plot')
  plt.xlabel(feature)
  plt.ylabel("Count")

  for bar in bars.patches:
    heigth = bar.get_height()
    plt.text(
            bar.get_x() + bar.get_width() / 2, heigth,
            f'{int(heigth)}',
            ha='center', va='bottom'
        )

  # rotate x-axis labels
  plt.xticks(rotation=45, ha='right')

plt.show()
```

## Data Preparation

### Data Imputation

**For categorical values**
- Conditions and Preciptype seems to have very similar values. Since "Preciptype" columns have lots of missing values, it is better to drop the "Preciptype columns".
- Since conditions at one time is very similar to the condition before it. It is better to impute the data using the values before it.

**For continuous features**
It is better to use interpolation since there is dependency of weather variables along time. But for simplicity, we can use the non-missing values before it.

- Severrisk: Severrisk is the probability of high risk weather. Upon further analysis, we can impute the data for severrisk using 0 or mean value.


- Drop windgust due to very large number of missing values
- Using the data imputation, we are going to prepare our data for further processing

```python
drop_columns = ["preciptype", "windgust"]
train_dropped_data = train_data.drop(labels = drop_columns, axis = 1)
train_dropped_data.head()
```

```python
train_dropped_data.isnull().sum()
```

```python
# first fill the severrisk data with
train_dropped_data["severerisk"].fillna(train_dropped_data["severerisk"].mean(), inplace = True)
```

```python
columns_to_interpolate = ["Temperature", "feelslike", "dewpoint", "humidity", "precipitation", "precipprob", "snow", "snowdepth", "windspeed", "winddirection", "sealevelpressure", "cloudcover", "visibility", "solarradiation", "uvindex"]
train_dropped_data[columns_to_interpolate].head(10)
```

```python
train_dropped_data[columns_to_interpolate] = train_dropped_data[columns_to_interpolate].fillna(method='ffill').fillna(method='bfill')
```

Hence, our training data is completed and now save it

But before we do that, we still have to perfor one hot encoding for the categorical columns

```python
# we have already dropped the preciptype since it is very similar to the weather conditions
# also preciptype has large number of missing values
# use data imputation in the "conditions" columns
train_dropped_data["conditions"] = train_dropped_data["conditions"].fillna(method="ffill")
```

```python
train_dropped_data["Name"] = train_dropped_data["Name"].fillna("Pokhara")
```

```python
train_dropped_data.isnull().sum()
```

```python
# lets perform one hot encoding
train_dropped_data["conditions"] = train_dropped_data["conditions"].apply(
    lambda x: [condition.strip() for condition in x.split(",")]
)
```

```python
train_dropped_data["conditions"].head()
```

```python
train_dropped_data.head()
```

```python
# let's perform one hot encoding using the multilabel binarizer
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()

one_hot_encoded_df = pd.DataFrame(mlb.fit_transform(train_dropped_data["conditions"]), columns = mlb.classes_, index = train_dropped_data.index)
final_data = pd.concat([train_dropped_data, one_hot_encoded_df], axis = 1)
final_data.head()
```

```python
final_data.columns
```

```python
# now drop the conditions column
final_data = final_data.drop(["conditions"], axis = 1)
final_data.head()
```

```python
final_data.isnull().sum()
```

```python
final_data.to_csv(data_dir / "final_data.csv", index = False)
```
