```python
def seasonal_plot(X, y, period, freq, ax=None):
    if ax is None:
        _, ax = plt.subplots()
    palette = sns.color_palette("husl", n_colors=X[period].nunique(),)
    ax = sns.lineplot(
        x=freq,
        y=y,
        hue=period,
        data=X,
        ax=ax,
        palette=palette,
        legend=False,
        errorbar=('ci', 0)
    )
    ax.set_title(f"Seasonal Plot ({period}/{freq})")
    for line, name in zip(ax.lines, X[period].unique()):
        y_ = line.get_ydata()[-1]
        ax.annotate(
            name,
            xy=(1, y_),
            xytext=(6, 0),
            color=line.get_color(),
            xycoords=ax.get_yaxis_transform(),
            textcoords="offset points",
            size=14,
            va="center",
        )
    return ax
```

```python
X = train_data.copy()

# days within a week
X["day"] = X.index.dayofweek  # the x-axis (freq)
X["week"] = X.index.week  # the seasonal period (period)
seasonal_plot(X, y="Demand (MW)", period="week", freq="day")
```

```python
# days within a year
X["month"] = X.index.month
X["year"] = X.index.year

seasonal_plot(X, y="Demand (MW)", period="year", freq="month");
```

```python
X["hour"] = X.index.hour
X["day"] = X.index.day

seasonal_plot(X, y="Demand (MW)", period="day", freq="hour");
```

## Periodogram

```python
def plot_periodogram(ts, detrend='linear', ax=None):
    from scipy.signal import periodogram
    fs = pd.Timedelta("365D") / pd.Timedelta("1H")
    freqencies, spectrum = periodogram(
        ts,
        fs=fs,
        detrend=detrend,
        window="boxcar",
        scaling='spectrum',
    )
    if ax is None:
        _, ax = plt.subplots()
    ax.step(freqencies, spectrum, color="purple")
    ax.set_xscale("log")
    ax.set_xticks([1, 4, 12, 52, 365, 8760])
    ax.set_xticklabels(
        [
            "Annual (1)",
            "Quarterly (4)",
            "Monthly (12)",
            "Weekly (52)",
            "Daily (365)",
            "Hourly (8760)"
        ],
        rotation=30,
    )
    ax.ticklabel_format(axis="y", style="sci", scilimits=(0, 0))
    ax.set_ylabel("Variance")
    ax.set_title("Periodogram plot for energy demand")
    return ax
```

```python
plot_periodogram(train_data["Demand (MW)"])
```

```python
annual_fourier = CalendarFourier(freq="A", order=4) # 10 sin/cos pairs for annual seasonality
daily_fourier = CalendarFourier(freq="D", order=2) # 2 sin/cos pairs for daily fourier

dp = DeterministicProcess(
    index = train_data.index,
    constant = True,
    order = 3,
    seasonal = False,
    additional_terms = [annual_fourier, daily_fourier]
)

X = dp.in_sample()
X.head()
```

```python
X.columns
```

```python
y = train_data["Demand (MW)"]

model = LinearRegression(fit_intercept = False)
_ = model.fit(X, y)

y_pred = pd.Series(model.predict(X), index = y.index)
X_fore = dp.out_of_sample(steps=168)
y_fore = pd.Series(model.predict(X_fore), index = X_fore.index)
```

```python
# axis to plot
ax = y["2023" : ].plot(color="0.25", style=".", title = "Demand (MW)")
ax = y_pred["2023" : ].plot(ax = ax, label="seasonal")
ax = y_fore["2023" : ].plot(ax=ax, label = "seasonal forecast", color="C3")
_ = ax.legend
```
